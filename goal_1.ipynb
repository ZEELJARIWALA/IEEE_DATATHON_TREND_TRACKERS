```python
df = pd.read_csv("/content/drive/MyDrive/IEEE (EV PROBLEM)/DATA_SET/Electric_Vehicle_Population_Data.csv")
```

#EV Adoption by City


```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import plotly.express as px

# Group data by City and count EVs
ev_adoption_by_city = df.groupby("City")["VIN (1-10)"].count().reset_index()
ev_adoption_by_city.columns = ["City", "EV Count"]
ev_adoption_by_city = ev_adoption_by_city.sort_values(by="EV Count", ascending=False)

# Select top 10 cities
top_10_cities = ev_adoption_by_city[:10]

# Create an interactive bar chart
fig = px.bar(
    top_10_cities,
    x="City",
    y="EV Count",
    title="Top 10 Cities by EV Adoption",
    labels={"EV Count": "Number of EVs", "City": "City"},
    text="EV Count",
    template="plotly_white"
)

fig.update_traces(textposition="outside")
fig.update_layout(xaxis_tickangle=-45)

# Show the interactive plot
fig.show()
```

# EV Type Distribution (Battery Electric Vehicles (BEV) vs Plug-in Hybrid Electric Vehicles (PHEV))

```python
import plotly.express as px

# Count occurrences of each EV Type
ev_type_distribution = df["Electric Vehicle Type"].value_counts().reset_index()
ev_type_distribution.columns = ["EV Type", "Count"]

# Create an interactive pie chart
fig = px.pie(
    ev_type_distribution,
    names="EV Type",
    values="Count",
    title="EV Type Distribution (BEV vs PHEV)",
    hole=0.3,  # Creates a donut-style chart
    template="plotly_white"
)

# Show the interactive plot
fig.show()
```

# EV Adoption Over the Years


```python
import plotly.express as px

# Group data by Model Year and count EVs
ev_adoption_by_year = df.groupby("Model Year")["VIN (1-10)"].count().reset_index()
ev_adoption_by_year.columns = ["Year", "EV Count"]

# Create an interactive line chart
fig = px.line(
    ev_adoption_by_year,
    x="Year",
    y="EV Count",
    title="EV Adoption Trend Over the Years",
    markers=True,
    labels={"EV Count": "Number of EVs", "Year": "Year"},
    template="plotly_white"
)

fig.update_traces(line=dict(width=3))  # Adjust line width for better visibility
fig.update_layout(xaxis=dict(tickmode="linear"))  # Ensure all years are shown

# Show the interactive plot
fig.show()
```
